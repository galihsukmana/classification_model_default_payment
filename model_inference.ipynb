## Model Inference

```python
# Import Library

import pickle
import joblib
import json
import pandas as pd
import numpy as np
```

## Load Model

```python

with open('model_best.pkl', 'rb') as file_1:
  model_best = pickle.load(file_1)

with open('model_scaler.pkl', 'rb') as file_2:
  model_scaler = pickle.load(file_2)

with open('model_encoder.pkl', 'rb') as file_3:
  model_encoder = pickle.load(file_3)

with open('list_num_cols.txt', 'r') as file_4:
  list_num_cols = json.load(file_4)

with open('list_cat_cols.txt','r') as file_6: 
  list_cat_cols = json.load(file_6)
```

## Inferencing

```python
df = pd.read_csv('data_inf.csv')
inf = df.tail(5).reset_index().drop(['index','Unnamed: 0'],axis=1)
inf
```

```python
# Split between Numerical Columns and Categorical Columns

data_inf_num = inf[list_num_cols]
data_inf_cat = inf[list_cat_cols]
data_inf_num
```

```python
# Feature Scaling and Feature Encoding
data_inf_num_scaled = model_scaler.transform(data_inf_num)
data_inf_num_scaled_df = pd.DataFrame(data_inf_num_scaled, columns=list_num_cols)
```

```python
data_inf_cat_encoded = model_encoder.fit_transform(data_inf_cat[['sex','marital_status']])
data_inf_cat_encoded_df = pd.DataFrame(data_inf_cat_encoded, columns=model_encoder.get_feature_names(data_inf_cat.columns))
data_inf_cat_encoded_df
```

```python
data_inf_final = pd.concat([data_inf_num_scaled_df, data_inf_cat_encoded_df], axis=1)
data_inf_final
```

```python
#predict using logistic regression
y_pred_inf = model_best.predict(data_inf_final)
y_pred_inf
```
